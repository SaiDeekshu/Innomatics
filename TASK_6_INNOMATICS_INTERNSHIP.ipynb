### QUESTION 1: Smart Parking Lot Management System

```python
def parking_system():
    capacity = int(input("Enter parking capacity: "))
    logs = list(map(str, input("Enter vehicle logs (IN/OUT): ").split()))
    count = 0
    max_count = 0
    for move in logs:
        if move == "IN":
            count += 1
        elif move == "OUT":
            count -= 1
        if count > max_count:
            max_count = count
    print("Currently Parked Vehicles:", count)
    if count < capacity:
        free = capacity - count
        print("Parking Status: Available")
        print("Free Spaces:", free)
        print("Vehicles that can still enter:", free)
    elif count == capacity:
        print("Parking Status: Full")
        print("No space available")
    else:   
        extra = count - capacity
        print("Parking Status: Overloaded")
        print("Extra Vehicles Inside:", extra)
        print("Vehicles that must exit:", extra)
    if max_count > capacity:
        exceed = max_count - capacity
        print("Alert: Parking Exceeded Capacity")
        print("Maximum Extra Vehicles at Peak Time:", exceed)
parking_system()
```

### QUESTION 2: Online Food Delivery Time Estimator

```python
def delivery_time():
    while True:
        distance = int(input("Enter distance (km): "))
        traffic = input("Enter traffic level (Low/Medium/High): ")
        weather = input("Enter weather (Clear/Rainy): ")
        time = distance * 5   
        if traffic == "Medium":
            time = time + 10
        elif traffic == "High":
            time = time + 20
        if weather == "Rainy":
            time = time + 15
        print("Estimated Delivery Time:", time, "minutes")
        choice = input("Do you want to calculate again? (yes/no): ")
        if choice == "no":
            print("Thank you!")
            break
delivery_time()
```

### QUESTION 3: Movie Theatre Seat Occupancy Analyzer

```python
def seat_analysis():
    total_seats = int(input("Enter total seats: "))
    booked = list(map(int, input("Enter booking list (1=booked): ").split()))
    booked_count = 0
    for seat in booked:
        if seat == 1:
            booked_count += 1
    percent = (booked_count / total_seats) * 100
    print("Occupancy:", percent, "%")
    if percent == 100:
        print("Show Status: Housefull")
    elif percent >= 70:
        print(f"Show Status: Almost Full and left seat is {100-percent}")
    else:
        print(f"Show Status: Available and left seat is {100-percent}")
seat_analysis()
```

### QUESTION 4: Cloud Server Load Classification System

```python
def server_check():
    cpu = list(map(int, input("Enter CPU readings: ").split()))
    total = 0
    count = 0
    for x in cpu:
        total += x
        count += 1
    avg = total / count
    print("Average CPU Load:", avg, "%")
    if avg < 50:
        print("Server Status: Normal")
    elif avg <= 80:
        print("Server Status: Warning")
    else:
        print("Server Status: Critical")
server_check()
```

### QUESTION 5: Smart Classroom Resource Usage Monitor

```python
def classroom_monitor():
    resources = {}
    n = int(input("Enter number of resources: "))
    for i in range(n):
        name = input("Enter resource name: ")
        hours = int(input("Enter usage hours: "))
        resources[name] = hours
    overused = []
    for item in resources:
        if resources[item] > 7:
            overused.append(item)
    if len(overused) > 0:
        print("Overused Resources:", overused)
        print("Energy Alert: Yes")
    else:
        print("Overused Resources: None")
        print("Energy Alert: No")
classroom_monitor()
```

### QUESTION 6: Online Event Registration Capacity Controller

```python
def event_control():
    capacity = int(input("Enter event capacity: "))
    registered = int(input("Enter total registrations: "))
    if registered <= capacity:
        confirmed = registered
        waitlist = 0
        status = "Open"
    else:
        confirmed = capacity
        waitlist = registered - capacity
        status = "Closed"
    print("Confirmed Registrations:", confirmed)
    print("Waitlisted Users:", waitlist)
    print("Registration Status:", status)
event_control()
```
