Problem 1: Unique Words in a Sentence

```python
sentence = input("Enter sentence: ")

#  Split sentence into words
words = sentence.split()

# set (removes duplicates)
unique_words = set(words)


print("Unique words count:", len(unique_words))
print("Unique words:", unique_words)
```

Problem 2: Highest Salary from Employee Data

```python
employees = {
    "Ravi": 75000,
    "Anita": 68000,
    "Kiran": 72000
}

#  Find maximum salary
highest_salary = max(employees.values())


for name, salary in employees.items():
    if salary == highest_salary:
        print("Highest Salary:", name, "-", salary)
```

Problem 3: Find Maximum and Minimum Values

```python
numbers = [45, 22, 89, 10, 66]

# first number as max and min
maximum = numbers[0]
minimum = numbers[0]


for num in numbers:
    if num > maximum:
        maximum = num
    if num < minimum:
        minimum = num

print("List:", numbers)
print("Max:", maximum)
print("Min:", minimum)
```

Problem 4: Count Products Above Price 1000

```python
prices = [450, 1200, 899, 1500, 300]

count = 0

for price in prices:
    if price > 1000:
        count += 1

print("Products above 1000:", count)
```

Problem 5: Calculate Attendance Percentage

```python
attendance = ["P", "P", "A", "P", "P"]

total_days = len(attendance)
present_days = attendance.count("P")

percentage = (present_days / total_days) * 100

print("Attendance Percentage:", percentage)
```

Problem 6: Remove Duplicate Phone Numbers

Problem 7: Count Character Frequency

```python
text = "python"

char_count = {}

for char in text:
    if char in char_count:
        char_count[char] += 1
    else:
        char_count[char] = 1

print(char_count)
```

Problem 10: Calculate Average Marks

```python
marks = [70, 75, 80, 65]

total = sum(marks)
count = len(marks)

average = total / count

print("Average Marks:", average)
```
