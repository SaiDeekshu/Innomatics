```python
!pip install langchain_community
from langchain_community.document_loaders import PyPDFLoader

file_name = list(uploaded.keys())[0]
loader = PyPDFLoader(file_name)

documents = loader.load()
print("Loaded pages:", len(documents))
```

```python
from langchain_text_splitters import RecursiveCharacterTextSplitter

splitter = RecursiveCharacterTextSplitter(
    chunk_size=500,
    chunk_overlap=50
)

chunks = splitter.split_documents(documents)

print("Total chunks:", len(chunks))
```

```python
from langchain_community.embeddings import HuggingFaceEmbeddings

embedding = HuggingFaceEmbeddings()
```

```python
from langchain_community.vectorstores import Chroma

vector_db = Chroma.from_documents(chunks, embedding)
```

```python
def generate_answer(query, docs):
    context = " ".join([d.page_content for d in docs])
    return f"Answer based on document:\n{context[:500]}"
```

```python
def process_query(query):
    docs = retrieve_docs(query)

    if len(docs) == 0:
        return "⚠️ Escalated to Human (No data found)"

    answer = generate_answer(query, docs)

    if len(answer) < 50:
        return "⚠️ Escalated to Human (Low confidence)"

    return answer
```

```python
def retrieve_docs(query):
    return vector_db.similarity_search(query, k=3)
```

```python
query = input("Ask your question: ")
print(process_query(query))
```

```python
from langgraph.graph import StateGraph
from typing import TypedDict

class State(TypedDict):
    query: str
    response: str

def process_node(state):
    query = state["query"]
    docs = retrieve_docs(query)

    if len(docs) == 0:
        state["response"] = "Escalated to Human"
    else:
        state["response"] = generate_answer(query, docs)

    return state

graph = StateGraph(State)
graph.add_node("process", process_node)
graph.set_entry_point("process")

app = graph.compile()

result = app.invoke({"query": "What is refund policy?"})
print(result["response"])
```
