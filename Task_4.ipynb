**Problem 1: Mobile Recharge Validation System**


```python
# List of valid recharge plans
VALID_PLANS = [199, 299, 399, 599]

# Function to validate recharge amount
def validate_recharge(amount):
    if amount < 50:
        print("âŒ Recharge amount must be at least â‚¹50.")
        return False

    if amount not in VALID_PLANS:
        print("âŒ Invalid plan selected.")
        print("Available plans are:", VALID_PLANS)
        return False

    return True


# Main program using while loop for retry
while True:
    try:
        recharge_amount = int(input("Enter recharge amount: â‚¹"))

        if validate_recharge(recharge_amount):
            print("âœ… Recharge successful!")
            break   # Exit loop if valid

        else:
            print("Please try again.\n")

    except ValueError:
        print("âŒ Please enter a valid number.\n")
```

**Problem 2: Inventory Reorder Alert System**

```python
# Function to check inventory status
def check_inventory(products):
    for product, quantity in products.items():
        if quantity < 15:
            print(f"âš ï¸ Reorder Alert: {product} has only {quantity} items left.")
        else:
            print(f"âœ… Stock OK: {product} has {quantity} items available.")


# Sample inventory dictionary
inventory = {
    "Rice": 20,
    "Oil": 10,
    "Sugar": 5,
    "Flour": 18,
    "Salt": 25
}

# Calling the function
check_inventory(inventory)
```

 **Problem 3: Student Result Processing System**

```python
# Function to process student result
def process_result(marks):
    total = 0

    # Loop to calculate total marks
    for mark in marks:
        total += mark

    # Calculate average
    average = total / len(marks)

    print(f"Average Marks: {average:.2f}")

    # Return result
    if average >= 50:
        return "Pass"
    else:
        return "Fail"


# Example list of student marks
student_marks = [60, 45, 70, 55, 80]

# Calling the function
result = process_result(student_marks)

print("Result:", result)
```

**Problem 4: Cab Fare Estimator with Retry Option**


```python
# Constants
BASE_FARE = 50
PER_KM_RATE = 12

# Function to calculate fare
def calculate_fare(distance, is_peak):
    fare = BASE_FARE + (distance * PER_KM_RATE)

    if is_peak:
        fare *= 1.25   # Adding 25% extra

    return fare


# Main program with retry option
while True:
    try:
        distance = float(input("Enter distance in km: "))

        peak_input = input("Is it peak hour? (yes/no): ").lower()
        is_peak = True if peak_input == "yes" else False

        total_fare = calculate_fare(distance, is_peak)

        print(f"ðŸš– Total Fare: â‚¹{total_fare:.2f}")

        retry = input("Do you want to calculate again? (yes/no): ").lower()
        if retry != "yes":
            print("Thank you for using Cab Fare Estimator!")
            break

    except ValueError:
        print("âŒ Please enter valid numeric input.\n")
```

**Problem 5: Employee Attendance Eligibility Checker**

```python
# Function to check attendance eligibility
def check_eligibility(attendance):
    present_count = 0

    # Loop to count present days
    for day in attendance:
        if day == "P":
            present_count += 1

    total_days = len(attendance)

    # Avoid division by zero
    if total_days == 0:
        return "No attendance data available"

    attendance_percentage = (present_count / total_days) * 100

    print(f"Attendance Percentage: {attendance_percentage:.2f}%")

    # Return eligibility status
    if attendance_percentage >= 75:
        return "Eligible"
    else:
        return "Not Eligible"


# Example attendance list
attendance_record = ["P", "P", "A", "P", "P", "A", "P", "P"]

# Calling function
result = check_eligibility(attendance_record)

print("Status:", result)
```

**Problem 6: Password Strength Checker**

```python
# Function to check password strength
def check_password_strength(password):

    # Rule 1: Minimum length
    if len(password) < 8:
        return "Weak Password (Minimum 8 characters required)"

    has_digit = False
    has_special = False

    # Loop over string
    for char in password:
        if char.isdigit():
            has_digit = True
        if char in "@#$":
            has_special = True

    # Check all conditions
    if has_digit and has_special:
        return "Strong Password âœ…"
    else:
        return "Weak Password (Must contain at least one digit and one special character @#$)"


# Example usage
password_input = input("Enter password: ")
result = check_password_strength(password_input)
print(result)
```
