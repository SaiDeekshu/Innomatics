**Problem statemnt 1: Employee performance eligibility**


```python
employees = {
    "Ravi": 92,
    "Anita": 88,
    "Kiran": 92,
    "Suresh": 85
}

#  Find highest score
highest_score = max(employees.values())

# Find all employees with highest score
top_performers = []

for name, score in employees.items():
    if score == highest_score:
        top_performers.append(name)

# Display result
print("Top Performers Eligible for Bonus:",
      ", ".join(top_performers),
      f"(Score: {highest_score})")
```

**Problem Statement 2: Search Query Keyword Analysis**

```python
import string

query = "Buy mobile phone buy phone online"

# Convert to lowercase
query = query.lower()

#  Remove punctuation
query = query.translate(str.maketrans('', '', string.punctuation))

#  Split into words
words = query.split()

#  Count frequency
word_count = {}

for word in words:
    if word in word_count:
        word_count[word] += 1
    else:
        word_count[word] = 1

#  Filter words searched more than once
result = {}

for word, count in word_count.items():
    if count > 1:
        result[word] = count

print(result)
```

**Problem Statement 3: Sensor Data Validation**


```python
sensor_readings = [3, 4, 7, 8, 10, 12, 5]

valid_readings = []

for hour, reading in enumerate(sensor_readings):
    if reading % 2 == 0:  # Check if even
        valid_readings.append((hour, reading))

print("Valid Sensor Readings (Hour, Value):")
print(valid_readings)
```

**Problem Statement 4: Email Domain Usage Analysis**


```python
emails = [
"ravi@gmail.com",
"anita@yahoo.com",
"kiran@gmail.com",
"suresh@gmail.com",
"meena@yahoo.com"
]

#  Total number of emails
total_emails = len(emails)

# Count domains
domain_count = {}

for email in emails:
    domain = email.split("@")[1]

    if domain in domain_count:
        domain_count[domain] += 1
    else:
        domain_count[domain] = 1

#  Calculate and display percentage
for domain, count in domain_count.items():
    percentage = (count / total_emails) * 100
    print(f"{domain}: {int(percentage)}%")
```

**Problem Statement 5: Sales Spike Detection**

```python
sales = [1200, 1500, 900, 2200, 1400, 3000]

#  Calculate average
total_sales = sum(sales)
average = total_sales / len(sales)

#  Calculate 30% above average threshold
threshold = average * 1.30

#  Detect spike days
for day, sale in enumerate(sales, start=1):
    if sale > threshold:
        print(f"Day {day}: {sale}")
```

**Problem Statement 6: Duplicate User ID Detection**


```python
user_ids = ["user1", "user2", "user1", "user3", "user1", "user3"]

#  Count occurrences
id_count = {}

for user in user_ids:
    if user in id_count:
        id_count[user] += 1
    else:
        id_count[user] = 1

#  Display duplicates only
for user, count in id_count.items():
    if count > 1:
        print(f"{user} → {count} times")
```
